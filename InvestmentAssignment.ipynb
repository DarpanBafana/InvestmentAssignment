```python
import numpy as np
import pandas as pd

companyFilePath = "companies.txt"
round2FilePath = "rounds2.csv"

companies = pd.read_csv(companyFilePath, sep="\t", encoding = "ISO-8859-1")
companies.describe()
cl = companies['permalink'].str.lower().unique().tolist()
print(len(cl))

rounds2 = pd.read_csv(round2FilePath, encoding = "ISO-8859-1")
r2l = rounds2['company_permalink'].str.lower().unique().tolist() 
print(len(r2l))
```

```python
companies['permalink'] = companies['permalink'].str.lower()
companies
```

```python
rounds2['permalink'] = rounds2['company_permalink'].str.lower()
rounds2
```

```python
master_frame = pd.merge(rounds2, companies, how='inner', on=['permalink'])
master_frame
```

#### Data Cleaning Steps

```python
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
import numpy as np

# removing rows where raised_amount_usd is NaN
master_frame = master_frame[~np.isnan(master_frame['raised_amount_usd'])]

round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
#removing unnecessary columns
master_frame = master_frame.drop('funding_round_code', axis=1)
master_frame = master_frame.drop('company_permalink', axis=1)
master_frame = master_frame.drop('founded_at', axis=1)

round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
def filterCategory(value):
    values = value.split('|')
    if len(values) > 1:
        return values[0]
    else:
        return value
```

```python
print(filterCategory('darpan|bafana'))
```

```python
top9
```

```python
# it was giving error - cannot apply split on FLOAT value so handled using removing inappropriate category_list rows
top9 = top9.dropna(axis=0, subset=['category_list'])
top9.info()
top9['primary_sector'] = top9['category_list'].apply(filterCategory)
top9
```

```python
top9['primary_sector'] = top9['primary_sector'].str.lower()
top9
```

```python
top9.groupby('country_code')['raised_amount_usd'].sum().sort_values(ascending=False)
```

```python
# pick top 3 english speaking countries
top3_english = top9[top9['country_code'].isin(['USA', 'GBR', 'IND'])]
top3_english.groupby('country_code')['raised_amount_usd'].sum().sort_values(ascending=False)
```

#### Mapping

```python
mapping = "mapping.csv"
mapping = pd.read_csv(mapping, encoding = "ISO-8859-1")
mapping
```

```python
mapping = pd.melt(mapping,id_vars=["category_list"])
mapping = mapping[mapping.value != 0]
mapping
```

```python
mapping = mapping.drop('value', axis=1)
mapping.rename(columns={'variable':'main_sector','category_list':'primary_sector'}, inplace=True)
mapping['primary_sector'] = mapping['primary_sector'].str.lower()
mapping
```

```python
top3_english
```

```python
top3_english = pd.merge(top3_english, mapping, how='left', on='primary_sector', indicator=True)
top3_english
```

```python
left_only = top3_english[top3_english['_merge'] != 'both']
left_only.primary_sector.sort_values(ascending=True).unique()
left_only
```

```python
top3_english = top3_english[top3_english['_merge'] == 'both']
top3_english
```

```python
top3_english = top3_english.drop('_merge', axis=1)
top3_english
```

```python
top3_english = top3_english.drop(top3_english[(top3_english.raised_amount_usd < 5000000)].index)
top3_english = top3_english.drop(top3_english[(top3_english.raised_amount_usd > 15000000)].index)
top3_english
```
